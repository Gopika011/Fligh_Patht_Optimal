```python
import matplotlib.pyplot as plt
import seaborn as sns

xf=out_df.drop(["timestamp","filename","lat","lon","waypoint_ID","routeId","altitude","speed","direction","processing_error","wind_degree"],axis=1)
print(xf)
def find_covariance(X):
    # covariance
    X_mean = X.mean()

# Standard deviation
    X_std = X.std()

# Standardization
    Z = (X - X_mean) / X_std
    c = Z.cov()

# Plot the covariance matrix

    sns.heatmap(c)
    plt.show()
    
    eigenvalues, eigenvectors = np.linalg.eig(c)
   
    # Index the eigenvalues in descending order 
    idx = eigenvalues.argsort()[::-1]

# Sort the eigenvalues in descending order 
    eigenvalues = eigenvalues[idx]

# sort the corresponding eigenvectors accordingly
    eigenvectors = eigenvectors[:,idx]
    explained_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)
    n_components = np.argmax(explained_var >= 0.50) + 1
    print(n_components)
find_covariance(xf)
    
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import numpy as np
from kneed import KneeLocator


def find_optimal_clusters(wp,X_train, max_k=11):
    
    
    scores = []
    silhouette_scores = {}
    for k in range(2, max_k):
        kmeans = KMeans(n_clusters=k,random_state=42)
        kmeans_labels = kmeans.fit_predict(X_train)
        scores.append(kmeans.inertia_)
        silhouette_avg = silhouette_score(X_train, kmeans_labels)
        silhouette_scores[k]=silhouette_avg
        
    #plt.figure(figsize=(10, 6))
    #plt.plot(range(2, 11), scores, marker='X', linestyle='--')
    #plt.title('Elbow Method')
    #plt.xlabel('Number of clusters')
    #plt.ylabel('WCSS')
   
   # plt.show()
    max=0
    n=0
    for key, value in silhouette_scores.items():
        if value>max:
            max=value
            n=key
    kl = KneeLocator(range(2, max_k), scores, curve="convex", direction="decreasing")
   
    return kl.elbow
    
from joblib import dump   
from sklearn.decomposition import PCA   
old_data=[]
def form_clusters(wp,data):
    #features = [ "temp_c" , "pressure","humidity",
     #             "wind_speed", "wind_sin","wind_cos", "clouds"]
    
    
    n_data=data[features]
   
    print(n_data)
    
    #delete data in folders
    if os.path.exists(f"./output/jup/waypoint_{wp}.csv"):
        os.remove(f"./output/jup/waypoint_{wp}.csv")
    if os.path.exists(f"./scalers/jup/scaler_{wp}.joblib"):
        os.remove(f"./scalers/jup/scaler_{wp}.joblib")
    if os.path.exists(f"./pca/jup/pca_{wp}.joblib"):
        os.remove(f"./pca/jup/pca_{wp}.joblib")
    if os.path.exists(f"./models/jup/waypoint_{wp}.joblib"):
        os.remove(f"./models/jup/waypoint_{wp}.joblib")
    #else create folders
    if not os.path.exists(f"./output/jup"):
        os.makedirs(f"./output/jup")
    if not os.path.exists(f"./scalers/jup"):
        os.makedirs(f"./scalers/jup")
    if not os.path.exists(f"./pca/jup"):
        os.makedirs(f"./pca/jup")
    if not os.path.exists(f"./models/jup"):
        os.makedirs(f"./models/jup")
    
    
    scaler = StandardScaler()
   
    scaled_data = scaler.fit_transform(n_data.values)
    print(scaled_data)
    pca = PCA(n_components=2)
    reduced_data=pca.fit_transform(scaled_data)
    #print(reduced_data)
    
    dump(scaler, f'./scalers/jup/scaler_{wp}.joblib')
    dump(pca, f'./pca/jup/pca_{wp}.joblib')
    
    n=find_optimal_clusters(wp,reduced_data)
    kmeans = KMeans(n_clusters=n)
    kmeans_labels = kmeans.fit_predict(reduced_data)
   
    dump(kmeans, f'./models/jup/waypoint_{wp}.joblib')
    data["cluster"]=kmeans_labels
    data.to_csv(f"./output/jup/waypoint_{wp}.csv")
    
    #represent visually
    
    
    
   

        
```

```python
import time
import requests
import pandas as pd
import json
from pprint import pprint
import weatherapi
def chunk_locations(locations, chunk_size=50):
    """Splits the locations list into smaller chunks of size 'chunk_size'."""
    return [locations[i:i + chunk_size] for i in range(0, len(locations), chunk_size)]
def fetch_weather(lat=0, lon=0, timestamp=None, max_retries=3, delay=2,locations=None):
    if timestamp:
        API_KEY = '9e195d193c40d821c3f6e18c693a1fd8'
        url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_KEY}"
        
        url += f"&dt={timestamp}"  # Add timestamp for historical data
    
        for attempt in range(max_retries):
            try:
                response = requests.get(url)
                response.raise_for_status()  # Raise error for bad responses
                data = response.json()
    
                if "main" in data and "wind" in data and "clouds" in data:
                    return data  # Return valid data
    
                print(f" Attempt {attempt + 1}: No valid data for ({lat}, {lon}). Retrying...")
            except requests.exceptions.RequestException as e:
                print(f" Attempt {attempt + 1}: Error fetching data for ({lat}, {lon}) - {e}")
    
            time.sleep(delay)  # Wait before retrying
    
        print(f" Max retries reached for ({lat}, {lon}). Skipping...")
        return None
    else:   
        API_KEY = '6bc045cf8fee4d6abdc75759250403'
        BULK_ENDPOINT = f"http://api.weatherapi.com/v1/current.json?key={API_KEY}&q=bulk"
        headers = {"Content-Type": "application/json"}
        payload = {"locations": locations}
    
        for attempt in range(max_retries):
            try:
                response = requests.post(BULK_ENDPOINT, headers=headers, data=json.dumps(payload))
                if response.status_code >= 400:
                    try:
                        error_info = response.json().get("error", {})
                        error_code = error_info.get("code")
                        error_message = error_info.get("message")
                        print(f" Attempt {attempt + 1}: Error {response.status_code} (code {error_code}) - {error_message}")
                    except Exception:
                        print(f" Attempt {attempt + 1}: Error {response.status_code} - Unable to parse error details.")
                    response.raise_for_status()
                
                data = response.json()
                if "bulk" in data:
                    return data["bulk"]  # Return the bulk list from the response
                else:
                    print(f" Attempt {attempt + 1}: 'bulk' field not found in response. Retrying...")
            except requests.exceptions.RequestException as e:
                print(f" Attempt {attempt + 1}: Exception occurred - {e}")
            
            time.sleep(delay)
        
        print(" Max retries reached for bulk request. Skipping...")
        return None

        


standard_path=pd.read_csv("flightdata/average_route.csv")

locations = []
for idx, row in standard_path.iterrows():
    q = f"{row['latitude']},{row['longitude']}"
    custom_id = f"loc-{idx}"  # You can change this to use an existing column if desired
    locations.append({
        "q": q,
        "custom_id": custom_id
    })
print(locations)
location_chunks = chunk_locations(locations, chunk_size=50)
bulk_response = []
for i, chunk in enumerate(location_chunks, start=1):
    print(f"\n--- Processing chunk {i}/{len(location_chunks)} ---")
    response = fetch_weather(locations=chunk)
    pprint(response)
    if response:
        bulk_response.extend(response)
```

```python
from joblib import load
models = glob.glob("./models/jup/*.joblib")
scalers = glob.glob("./scalers/jup/*.joblib")
pcas = glob.glob("./pca/jup/*.joblib")

def filter_waypoints(df,cluster_num):
    routes=[]
    #debug
    print(df)
    
    for num,i in df.iterrows():
        
        if i["cluster"]==cluster_num:
            routes.append(i["routeId"])
    return routes
    
```

```python
wp_num = 12
print(features)
# Change this to view different waypoints
try:
    features.index("cluster")
except ValueError:
    features.append("cluster")
# Load training data and models
train_df = pd.read_csv(f"./output/jup/waypoint_{wp_num}.csv")
train_df=train_df[features]

scaler = load(f'./scalers/jup/scaler_{wp_num}.joblib')
pca = load(f'./pca/jup/pca_{wp_num}.joblib')
kmeans = load(f'./models/jup/waypoint_{wp_num}.joblib')

# Process current weather data
current_features = current_data[wp_num]

scaled_current = scaler.transform([current_features])
reduced_current = pca.transform(scaled_current)
cluster = kmeans.predict(reduced_current)[0]

# Visualize
plt.figure(figsize=(10, 6))
colors = plt.cm.tab10.colors  # Color map for clusters

# Plot training data clusters
for c in range(kmeans.n_clusters):
    cluster_data = train_df[train_df['cluster'] == c]
   
    reduced_train = pca.transform(scaler.transform(cluster_data.iloc[:, :-1].values))  # Exclude metadata columns
    plt.scatter(reduced_train[:, 0], reduced_train[:, 1],
                color=colors[c], label=f'Cluster {c}', alpha=0.6)

# Plot current prediction
plt.scatter(reduced_current[:, 0], reduced_current[:, 1],
            color='red', marker='*', s=300,
            label=f'Current (Cluster {cluster})', edgecolors='black')

plt.title(f'Waypoint {wp_num} Cluster Visualization\n(PCA Reduced Space)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load training data and models
wp_num =50  # Change this to view different waypoints
train_df = pd.read_csv(f"./output/jup/waypoint_{wp_num}.csv")
train_df = train_df[features]

scaler = load(f'./scalers/jup/scaler_{wp_num}.joblib')
pca = load(f'./pca/jup/pca_{wp_num}.joblib')
kmeans = load(f'./models/jup/waypoint_{wp_num}.joblib')

# Process current weather data
current_features = current_data[wp_num]
scaled_current = scaler.transform([current_features])
reduced_current = pca.transform(scaled_current)
cluster = kmeans.predict(reduced_current)[0]

# Add the current weather data to the training dataframe for visualization
current_df = pd.DataFrame([current_features], columns=features[:-1])  # Exclude 'cluster' column
current_df['cluster'] = cluster
train_df = pd.concat([train_df, current_df], ignore_index=True)

# Generate scatter plots for each feature combination
feature_combinations = [(features[i], features[j]) for i in range(len(features) - 1) for j in
                        range(i + 1, len(features) - 1)]

for feature_x, feature_y in feature_combinations:
    plt.figure(figsize=(10, 6))
    colors = plt.cm.tab10.colors  # Color map for clusters

    # Plot training data clusters
    for c in range(kmeans.n_clusters):
        cluster_data = train_df[train_df['cluster'] == c]
        plt.scatter(cluster_data[feature_x], cluster_data[feature_y],
                    color=colors[c], label=f'Cluster {c}', alpha=0.6)

    # Plot current prediction
    current_point = train_df.iloc[-1]  # Last row is the current weather data
    plt.scatter(current_point[feature_x], current_point[feature_y],
                color='red', marker='*', s=300,
                label=f'Current (Cluster {cluster})', edgecolors='black')

    plt.title(f'Waypoint {wp_num}: {feature_x} vs {feature_y}\n(Cluster Visualization)')
    plt.xlabel(feature_x)
    plt.ylabel(feature_y)
    plt.legend()
    plt.grid(True)
    plt.show()
```
